# Ridge Regression Gradient Descent.

##Libraries Imported

```python
from sklearn.datasets import load_diabetes
from sklearn.metrics import r2_score
import numpy as np
```

## Data Loading

```python
X,y = load_diabetes(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=4)
```

## Applying SGD Regressor

Note -> Because there is randomness in SGD The coefficient value Will not be same In multiple trainings.

```python
from sklearn.linear_model import SGDRegressor
```

```python
# here penalty = L2 -> L2 Norm.
reg = SGDRegressor(penalty='l2',max_iter = 500,eta0=0.1,learning_rate='constant',alpha=0.001)
```

```python
reg.fit(X_train,y_train)

y_pred = reg.predict(X_test)
print("R2 Score",r2_score(y_test,y_pred))
print(reg.coef_)
print(reg.intercept_)
```

## Applying Ridge GD for Coef.

Note -> In Ridge there's no randomness. That's why value of coefficient will always be same.

```python
from sklearn.linear_model import Ridge
```

```python
# Sparse_cg -> used as solver for GD.

reg = Ridge(alpha=0.001,max_iter=500,solver='sparse_cg')
```

```python
reg.fit(X_train,y_train)

y_pred = reg.predict(X_test)
print("R2 Score",r2_score(y_test,y_pred))
print(reg.coef_)
print(reg.intercept_)
```

## Manual Class for Ridge Regression GD

```python
class MeraRidge:

  def __init__(self,epochs,learning_rate,alpha):

    self.learning_rate = learning_rate
    self.epochs = epochs
    self.alpha = alpha
    self.coef_ = None
    self.intercept_ = None

  def fit(self,X_train,y_train):

    self.coef_ = np.ones(X_train.shape[1])
    self.intercept_ = 0
    thetha = np.insert(self.coef_,0,self.intercept_)

    X_train = np.insert(X_train,0,1,axis=1)

    for i in range(self.epochs):
      thetha_der = np.dot(X_train.T,X_train).dot(thetha) - np.dot(X_train.T,y_train) - self.alpha*thetha
      thetha = thetha - self.learning_rate*thetha_der




  def predict(self,X_test):
    return np.dot(X_test,self.coef_) + self.intercept_
```
